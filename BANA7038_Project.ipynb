```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("C:/Users/Reshma Punukollu/Downloads/Housing_Data.csv")
data=data.head(10)
data
```

```python
data.rename(columns={'No.of_Bedrooms':'Number_of_Bedrooms'},inplace=True)
data.rename(columns={'No.of_Parking_Spots':'Number_of_Parking_Spots'},inplace=True)
data.rename(columns={'No.of_Bathrooms':'Number_of_Bath_Rooms'},inplace=True)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

Year_Built = data['Year_Built']
Number_of_Parking_Spots = data['Number_of_Parking_Spots']

ax.scatter(Year_Built, Number_of_Parking_Spots , color = 'g')
ax.set_ylabel('Number_of_Parking_Spots')
ax.set_xlabel('Year_Built')
ax.set_title('Number of Parking Spots along with Year Built ')
ax.yaxis.set_major_formatter('{x:1.2f}')
ax.grid(linestyle = 'dashed')
plt.xticks(rotation = 90)
plt.show()
```

```python
##We performed an analysis of housing data in the greater Cincinnati area using the Python programming language. Specifically, we focused on the variables of Year Built and Number of Parking Spots to plot a trend of parking spots built in a given year. This allowed us to visualize the allocation of parking spaces for each house in a given year
```

```python
plot2 = sns.barplot(data = data, x = 'Number_of_Bath_Rooms', y = 'Number_of_Bedrooms', hue = 'Year_Built', orient = 'h')
plot2.set(xlabel = 'Number_of_Bath_Rooms', ylabel = 'Number_of_Bedrooms', title = 'Number of Bedrooms and Number of Bathrooms for a coressponding year')
plot2.legend(title = 'Year_Built', bbox_to_anchor=(1, 1), loc = 'upper left')
plot2.grid(axis = 'x')
```
